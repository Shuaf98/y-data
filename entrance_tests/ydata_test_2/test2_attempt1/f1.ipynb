### IMPORTANT NOTES: 1) Data types were changed in Excel. Make sure the conditions written here match the new data types. 2) Make sure to change index to False in the output line, as to not include the index as a column in the final answer.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\sfrie\Ydata_practice\Test_2\schools.csv")
df.head()
```

```python
df = df[df['Student Attendance Rate'] > 0]
df = df[df['Rigorous Instruction'] > 0]
df = df[df['Students Chronically Absent'] < 1]
df = df[(df['Student Achievement Rating'].notnull()) & (df['Average Proficiency Score'].notnull())]
```

```python
def convert(n):
    if n == 'Not Meeting Target':
        return 0.0
    elif n == 'Approaching Target':
        return 0.5
    elif n == "Meeting Target":
        return 0.8
    else:
        return 1.0
df['Student Achievement Rating'] = df['Student Achievement Rating'].apply(convert)
df
        
```

```python
df.to_csv(r"C:\Users\sfrie\Ydata_practice\Test_2\output.csv", index = False)
```
