```python
# To add a new cell, type '# %%'
# To add a new markdown cell, type '# %% [markdown]'
# %%
import pandas as pd
import numpy as np
import os
df = pd.read_csv('purchase_data.csv')
input = open('e1_input.txt').read()
prod1, prod2 = input.split()

members = list(df['memberID'].unique())

def find_common(prod1, prod2):
# %%
    all_members = []
    for x in members:
        test = df[df['memberID'] == x]
        if int(prod1) in test['productID'].values and int(prod2) in test['productID'].values:
            all_members.append(x)
    if all_members != []:
        return all_members
    else:
        return 0



# %%
```

```python
print(find_common(prod1, prod2))
```

```python
def similarity (product_1, product_2):  
    common = find_common(product_1, product_2) 
    sum_mult =0
    sum_sq_1=0
    sum_sq_2=0
    for member in common:  
        prod1_count = len(df[(df['memberID'] == member) & (df['productID'] == product_1)])
        prod2_count = len(df[(df['memberID'] == member) & (df['productID'] == product_2)])
        prod1_score = df[df['productID'] == int(product_1)]['productScore'].unique()
        prod2_score = df[df['productID'] == int(product_2)]['productScore'].unique()
        sum_mult += (prod1_count - prod1_score) *  (prod2_count - prod2_score)  
        sum_sq_1 += (prod1_count - prod2_count)^2  
        sum_sq_2 += (prod2_count - prod2_score)^2  
    similarity = sum_mult / (sqrt(sum_sq_1) * sqrt(sum_sq_2))
    print(similarity)
    return similarity
```
